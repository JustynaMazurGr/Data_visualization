ZADANIE 7. W problemie optymalizacji doboru strategii dla zestawu graczy
uruchomione zostały trzy algorytmy. Każdy z algorytmów wygenerował
łącznie 100 różnych rozwiązań. Kluczowe w rozwiązaniu jest największe
dostępne prawdopodobieństwo wyboru strategii (dane z pliku). Na
wykresie rozrzutu przedstaw wyniki z pliku. Zastosuj poniższe metody
poprawy czytelności wykresu:
faceting;
próbkowanie (wybierz próbkę wielkości ok 20 - 25% całego zestawu);
przedstaw wyniki na wykresie posługując się czterema różnymi
wielkościami punktów: najmniejsze – jeżeli prawd. wyboru jest
mniejsze od 0.25, większe – prawd. w przedziale (0.25, 0.5⟩, duże –
prawd. w przedziale (0.5, 0.75⟩, największe – prawd. powyżej 0.75.

```python
# ZADANIE 7

import pandas
import matplotlib.pyplot as plt

#wczytanie danych do data frame
path_1 = '/content/drive/MyDrive/wykresy zadania/set1.txt'
path_2 = '/content/drive/MyDrive/wykresy zadania/set2.txt'
path_3 = '/content/drive/MyDrive/wykresy zadania/set3.txt'

df_1 = pandas.read_csv(path_1, header=None)  #brak nagłówka
df_1 = df_1.T.head(25)                       #wybieram próbkę 25% całego zestawu, górne 25 elementów wiedząc, że dataFrame ma 100 elementów
df_2 = pandas.read_csv(path_2, header=None).head(25)
df_2 = df_2.T.head(25)
df_3 = pandas.read_csv(path_3, header=None).head(25)
df_3 = df_3.T.head(25)

x = range(len(df_1))
y1 = df_1[0]
y2 = df_2[0]
y3 = df_3[0]

#wielkości kropek
x_size1 = []
x_size2 = []
x_size3 = []
for i in range(0,len(df_1),1):
    if df_1[0][i] <= 0.25:  #na opak z powodu transpozycji
        x_size1.append(10)
    elif(df_1[0][i] > 0.25) and (df_1[0][i] <= 0.5):    #w przedziale (0.25, 0.5⟩
        x_size1.append(20)
    elif(df_1[0][i] > 0.5) and (df_1[0][i] <= 0.75):    #w przedziale (0.5, 0.75⟩
        x_size1.append(30)
    elif df_1[0][i] > 0.75:
        x_size1.append(40)

for i in range(0,len(df_2),1):
    if df_2[0][i] <= 0.25:  
        x_size2.append(10)
    elif(df_2[0][i] > 0.25) and (df_2[0][i] <= 0.5):    
        x_size2.append(20)
    elif(df_2[0][i] > 0.5) and (df_2[0][i] <= 0.75):    
        x_size2.append(30)
    elif df_2[0][i] > 0.75:
        x_size2.append(40)

for i in range(0,len(df_3),1):
    if df_3[0][i] <= 0.25:  
        x_size3.append(10)
    elif(df_3[0][i] > 0.25) and (df_3[0][i] <= 0.5):    
        x_size3.append(20)
    elif(df_3[0][i] > 0.5) and (df_3[0][i] <= 0.75):    
        x_size3.append(30)
    elif df_3[0][i] > 0.75:
        x_size3.append(40)

#wykresy rozrzutu, zastosowany faceting i ustawienie wielkości kropek
fig, ax = plt.subplots(1,3)

fig.suptitle('Wykresy rozrzutu przedstawiające wyniki z plików.', fontsize=10)

#dodajemy odstępy pomiędzy wykresami
fig.tight_layout()

ax[0].scatter(x,y1, s = x_size1)
ax[0].scatter(x,y2, color = 'gray', s = x_size2)
ax[0].scatter(x,y3, color = 'gray', s = x_size3)
ax[0].set_title('niebieski: set1' ,fontsize = 7)

ax[1].scatter(x,y1, color = 'gray', s = x_size1)
ax[1].scatter(x,y2, s = x_size2)
ax[1].scatter(x,y3, color = 'gray', s = x_size3)
ax[1].set_title('niebieski: set2' ,fontsize = 7)

ax[2].scatter(x,y1, color = 'gray', s = x_size1)
ax[2].scatter(x,y2, color = 'gray', s = x_size2)
ax[2].scatter(x,y3, s = x_size3)
ax[2].set_title('niebieski: set3' ,fontsize = 7)

#ustawiam opis osi tylko po zewnętrznej dla Whole weight i pośrodku środkowego wykresu dla Length (nie ma sensu powtarzać)
ax[1].set_xlabel('Kolejne rozwiązania (od 0 do 25)', fontsize=7)
ax[0].set_ylabel('Wyniki z pliku', fontsize=7)

#ustawiam rozmiar etykiet na osiach x i y
ax[0].tick_params(axis='both', labelsize=7)
ax[1].tick_params(axis='both', labelsize=7)
ax[2].tick_params(axis='both', labelsize=7)

plt.savefig('/content/drive/MyDrive/wykresy zadania/Wykresy rozrzutu dla zestawów danych z trzech algorytmów dotyczących problemu optymalizacji doboru strategii graczy', dpi=200)

plt.show()
```

```python
# ZADANIE 8

import numpy as np
import matplotlib.pyplot as plt
import math

#funkcja Alpejska "2"
def f(x, y):
    return np.cos(x)*x*(1+(np.cos(y)/2))

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

X, Y = np. meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='magma')    

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('wykres 3D - funkcja Alpejska "2"')

plt.savefig('/content/drive/MyDrive/wykresy zadania/Wykres 3D funkcja Alpejska', dpi=200)
plt.show()


#funkcja Schuberta
def f2(x2, y2):
    return ((np.cos(2*x2+1))+(2*np.cos(3*x2+2))+(3*np.cos(4*x2+3))+(4*np.cos(5*x2+4))+(5*np.cos(6*x2+5)))*((np.cos(2*y2+1))+(2*np.cos(3*y2+2))+(3*np.cos(4*y2+3))+(4*np.cos(5*y2+4))+(5*np.cos(6*y2+5)))

x2 = np.linspace(0,12,30)
y2 = np.linspace(0,12,30)


X2, Y2 = np. meshgrid(x2, y2)
Z2 = f2(X2, Y2)

fig2 = plt.figure()
ax2 = plt.axes(projection='3d')
#ax2.contour3D(X2, Y2, Z2, 70, cmap='binary')
ax2.plot_surface(X2, Y2, Z2, cmap='magma')  

ax2.set_xlabel('x2')
ax2.set_ylabel('y2')
ax2.set_zlabel('z2')
ax2.set_title('wykres 3D - funkcja Schuberta')

plt.savefig('/content/drive/MyDrive/wykresy zadania/Wykres 3D funkcja Schuberta', dpi=200)
plt.show()


#funkcja Rastrigina:
def f3(x3, y3):
    return (x3**2-np.cos(18*np.pi*x3))+(y3**2-np.cos(18*np.pi*y3))

x3 = np.linspace(-10,10,30)
y3 = np.linspace(-10,10,30)

X3, Y3 = np. meshgrid(x3, y3)
Z3 = f3(X3, Y3)

fig3 = plt.figure()
ax3 = plt.axes(projection='3d')

ax3.plot_surface(X3, Y3, Z3, cmap='magma')    

ax3.set_xlabel('x3')
ax3.set_ylabel('y3')
ax3.set_zlabel('z3')
ax3.set_title('wykres 3D - funkcja Rastrigina')

plt.savefig('/content/drive/MyDrive/wykresy zadania/Wykres 3D funkcja Rastrigina', dpi=200)
plt.show()
```
